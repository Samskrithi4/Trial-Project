```python
import os

directory = "output_files"
dict_list = []
for filename in os.listdir(directory):
    filepath = os.path.join(directory, filename)
    if os.path.isfile(filepath):
        print(filepath)
        with open(filepath, 'r') as file:
            count = 0
            lines = file.readlines()
            dict_list.append(lines[-2].strip())



print(dict_list)
len(dict_list)
```

```python
from collections import defaultdict
import ast

master_dict = defaultdict(int)

for item in dict_list:
    dictionary = ast.literal_eval(item)
    
    for key, value in dictionary.items():
        master_dict[key] += value
        
master_dict = dict(master_dict)
sorted_master_dict = sorted(master_dict.items(), key=lambda item: item[1], reverse=True)
sorted_master_dict = dict(sorted_master_dict)
# sorted_master_dict = dict(sorted_master_dict[:10])
print(sorted_master_dict)
len(sorted_master_dict)
```

```python
import matplotlib.pyplot as plt

labels = list(sorted_master_dict.keys())
sizes = list(sorted_master_dict.values())

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Languages')
plt.show()
```
